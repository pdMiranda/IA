```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.datasets import load_iris
```

```python
# Carregando o conjunto de dados Iris (substitua pelo seu próprio conjunto de dados)
data = load_iris()
X = data.data
```

```python
# Pré-processamento: Normalização
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Executando o K-means com 3 clusters (número de classes no conjunto de dados Iris)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
predicted_labels = kmeans.labels_
true_labels = data.target
```

```python
# Avaliando a qualidade dos agrupamentos usando Silhouette Score
silhouette_avg = silhouette_score(X_scaled, predicted_labels)
print(f"Silhouette Score: {silhouette_avg:.2f}")
```

```python
# Investigando instâncias agrupadas incorretamente
incorrectly_grouped = np.where(predicted_labels != true_labels)[0]
print(f"Instâncias agrupadas incorretamente: {incorrectly_grouped}")
```
